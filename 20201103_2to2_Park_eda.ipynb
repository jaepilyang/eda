### 텍스트 분석

```python
# 정규표현식
# []: 대괄호 내의 문자와 일치하는 패턴
# [가나다]: 가나다 중에 하나와 일치
# [가-힣]: 한글
# [0-9]: 숫자
# [a-zA-Z]: 영문자
# [^0-9]: 숫자를 포함하지 않는 패턴
# [^가-힣]: 한글을 포함하지 않는 패턴
    
# %d: 숫자
# %s: 공백문자
# %w: 영문자
    
# ^x: x로 시작하는 문자
# x$: x로 끝나는 패턴
# .x: 한 문자와 x로 끝나는 패턴
# x+: x가 한 번 이상 등장하는 패턴
# x?: x가 0번 또는 한 번
# x*: x가 0번 이상 등장하는 패턴
# x{n}: x가 n번 등장하는 패턴
# x{n:m}: x가 n번 이상 m번 이하 등장하는 패턴
```

```python
# 공원보유시설(운동시설)에서 각각의 농구시설이 몇 개가 있는지 확인
```

```python
df['공원보유시설(운동시설)'].unique().tolist()
```

```python
gym = str(df['공원보유시설(운동시설)'].unique().tolist())
gym
```

```python
gym = gym.replace('+', ',').replace("'", ',').replace(" ", ',').replace("/", ',')
gym[:1000]
```

```python
re.sub('\d','',gym)
```

```python
gym = re.sub('[\[\d\-nan\.\]]', '', gym)
```

```python
gym = re.sub(",{2}", ',', gym)
gym[:100]
```

```python
gym_count = pd.Series(gym.split(',')).value_counts()
```

```python
plt.figure(figsize=(15,5))
gym_count[:50].plot(kind='bar')
```

```python
word_draw=WordCloud(
    font_path = 'C://Windows/Fonts/HMFMMUEX.TTC',
    width=1200,
    height=500,
    stopwords = ['운동기구', '개소', '외종'],
    background_color='white'
)
word_draw.generate(gym)

plt.figure(figsize=(30,15))
plt.imshow(word_draw)
plt.axis('off')
plt.show()
```
