```python
# 2010 ~ 2019 외국인 출입파일
# 전체 파일을 위의 전처리 후 하나의 df 통합
```

```python
def kto_df(date):
    df = pd.read_excel('./tourist_files/kto_{}.xlsx'.format(date), header=1, usecols = 'A:G', skipfooter=4)
    df['기준년월'] = date

    continent_list = ['아시아주', '미주', '구주', '대양주', '아프리카주', '기타대륙', '교포소계', '아프리카 기타']
    
    condition = (df['국적'].isin(continent_list) == False)
    country_df = df[condition]

    country_df.reset_index(drop=True, inplace=True)

    continents = ['아시아']*25 + ['아프리카']*5 + ['유럽']*23 + ['오세아니아']*3 + ['아프리카'] + ['기타대륙'] + ['교포']

    country_df['대륙'] = continents
    country_df['관광객비율'] = country_df['관광']/country_df['관광'].sum()
    return country_df
```

```python
date = '201905'
kto_df(date)
```

```python
kto_dfs = []

for y in range(2010, 2020):
    for m in range(1, 13):
        try:
            date = str(y)+str(m).zfill(2)
            df = kto_df(date)
            kto_dfs.append(df)
        except:
            pass

kdf = pd.concat(kto_dfs)
kdf
```
